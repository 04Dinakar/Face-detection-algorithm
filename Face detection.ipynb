```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("./grp.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
detector=cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
allfaces=detector.detectMultiScale(img,1.3,1)
allfaces.shape
```

```python
x,y,w,h=allfaces[0]
img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),1)
plt.imshow(img)
plt.show()
```

```python
for face in allfaces:
    x,y,w,h=face
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),2)
plt.imshow(img)
plt.show()
    
```

```python
detector=cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
cam=cv2.VideoCapture(0)
```

```python
while True:
    ret,frame=cam.read()
    if ret==False:
        continue
    allfaces=detector.detectMultiScale(frame,1.3,5)
    for face in allfaces:
        x,y,w,h=face
        frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,0),2)
    cv2.imshow("ok",frame)
    key=cv2.waitKey(1)&0xFF
    if key==ord('q'):
        break
cam.release()
cv2.destroyAllWindows()
    
```
